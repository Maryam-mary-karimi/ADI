```python
def decision_tree(permission, merit, afford, lifetime, update, late, trust, redund):    
    action="drop"

    if (permission=="no"):
        action=0
    else:
        if (merit=="low"):
            if(lifetime=="short"):
                action=1
            else:
                if(afford=="none"):
                    action=1
                else:
                    if(late=="low"):
                        action=2
                    else:
                        if(trust=="high"):
                            action=2
                        else: action=3
        else: 
            if(update=="yes"):
                if(afford=="none"):
                    action=1
                else:
                    if(lifetime=="short"):
                        if(trust=="hight"):
                            action=2
                        else:
                            action=3
                    else:
                        action=3
            else:
                if(afford=="none"):
                    action=1
                elif(afford=="some"):
                    action=3
                else:
                    if(trust=="high"):
                        if(redund=="yes"):
                            action=3
                        else:
                            action=4
                    else:
                        action=4

    return action;
```

```python
#fill ADI with verification method based on df characteristics
ADI.loc[:, ['Unnamed: 0']]=0
ADI.loc[:, ['inpatient.number']]=0
ADI.loc[:, ['admission.ward']]=1
ADI['occupation'] = ADI['occupation'].map({'UrbanResident': 'some', 'farmer': 'none', 'worker':'none', 'Officer':'extra','Others':'none', 'nan':'none'})
ADI=ADI.rename(columns = {'occupation':'affordability'})
ADI['admission.way'] = ADI['admission.way'].map({'NonEmergency': 'high', 'Emergency': 'low'})
ADI=ADI.rename(columns = {'admission.way':'LatencyTolerance'})

ADI['update']=ADI['DestinationDischarge']
ADI['update'] = ADI['update'].map({'Died': 'no'}).fillna('yes')

for r,row in ADI.iterrows():
    if(row[2]=='Home' or row[2]=='Unknown' ):
        ADI.at[r,'DestinationDischarge'] = 0
    if(row[2]=='Died'):
        ADI.at[r,'DestinationDischarge'] = 1
    if(row[2]=='HealthcareFacility'):
        #permission, merit, afford, lifetime, update, late, trust, redund
        ADI.at[r,'DestinationDischarge']=decision_tree('yes', "high",ADI.at[r,'affordability'], "short", ADI.at[r,'update'], ADI.at[r,'LatencyTolerance'], "yes", "no")

for r,row in ADI.iterrows():#permission, merit, afford, lifetime, update, late, trust, redund
    ADI.at[r,'discharge.department']=decision_tree('yes', "low",ADI.at[r,'affordability'], "short", "no", ADI.at[r,'LatencyTolerance'], "yes", "no")

ADI.head()
```
